```python
"""Case study and phase space modelling for 
Lotka Volterra differentioa equations"""
import numpy as np
import matplotlib.pyplot as plt
#set global variables
depth = 1024                     #overused, change breaks program, def.v=128
values = np.zeros((2,depth))
param = np.zeros((6,depth))
param[0,:],param[1,:]=np.linspace(0,3,depth),np.linspace(0,3,depth)
param[2,:],param[3,:]=np.linspace(1,150,depth),np.linspace(1,151,depth)
param[4,:],param[5,:]=np.linspace(0,3,depth),np.linspace(0,3,depth)
```

```python
"""Define function to calculate values"""
def update(j):
    """
    r1,r2 = 1.606,3.185               #0,87 - 2.37
    k1,k2 = 128.5,  90.1
    c1,c2 = 0.35, 0.6990
    """
    values[0,0] = 50 #for species1
    values[1,0] = 50 #for species2
    r1,r2 = param[0,240],param[1,512]
    k1,k2 = param[2,860],param[3,512]
    c1,c2 = param[4,540],param[5,512]
    
    for i in range(depth-1):
        n1 = values[0,i]
        n2 = values[1,i]
        dn1dt = n1*r1*(k1-n1-c2*n2)/k1
        dn2dt = n2*r2*(k2-n2-c1*n1)/k2
        values[0,i+1]=n1+dn1dt
        values[1,i+1]=n2+dn2dt
    print('r1=',r1,'r2=',r2,'k1=',k1,'k2=',k2,'c1=',c1,'c2=',c2)
    return values
```

```python
"""Generate sequences of frames with changing variable p
   plot functions on one graph
   phase space on another"""

size=15
num_of_slides = 1
for j in range(num_of_slides):
    
    update(j)
    print(j)
    plt.figure(figsize=(2*size,size))
    plt.plot(values[0,])
    plt.plot(values[1,])
    plt.show()
    
    #plt.figure(figsize=(2*size,size))
    plt.plot(values[0,],values[1,],'.g')
    plt.show()
    
```
